```python
n = X.shape[1]
alpha = 1e-4 # 0.1 0.01
W = np.array([1, 0.5])

W, alpha
```

```python
for i in range(1500):
    y_pred = np.dot(W, X)
    err = dal.calc_mse(y, y_pred)
    for ii in range(W.shape[0]):
        W[ii] -= alpha * (1/n * 2 * np.sum(X[ii]*(y_pred-y)))
    if i % 100 == 0:
        print(i, W, err)
```
